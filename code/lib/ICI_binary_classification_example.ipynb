# Using ICI for binary classification problem.

## Importing

```python
import numpy as np
from ICI import ICIHarmonization
```

# Generate data

```python
X = np.random.randn(200, 5)
y = np.array([0] * 140 + [1] * 60)
# randomize y
y = np.random.permutation(y)
sites = np.array([0] * 100 + [1] * 100)
```

## Let's create a 

```python
ici = ICIHarmonization("smote", verbose=True)
X_r, y_r = ici.fit_resample(X, y, sites=sites)

print(np.unique(y_r, return_counts=True))
```
